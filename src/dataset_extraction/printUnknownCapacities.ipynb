# Prints the names of all the cables with unknown capacity

```python
import networkx as nx
from pathlib import Path
import ipynbname#necessary to get filepath to jupyter notebook file

_FILE_DIR = ipynbname.path().parent.parent#obtain directory of this file
_PROJ_DIR = _FILE_DIR.parent#obtain main project directory
_EXTRACT_DIR = _PROJ_DIR / "dataExtracted"

G = nx.read_graphml(str(_EXTRACT_DIR / "THE_GRAPH.graphml"))
print(G)
```

```python
cableNames = set()#set that will contain all the cable names for which we have 0 capacities -> unknown!
for e in G.edges(data=True):
    data = e[2]#get the data part of the edge
    if data["capacity"] == 0:
        cableNames.add(data["name"])
        
print(len(cableNames))
for c in cableNames:
    print(c)
```
